# Immune Cell Database Design

## Overview
Design for a database to store immune cell count data from clinical studies.

## Entity-Relationship Diagram Concept

```mermaid
erDiagram
    PROJECTS ||--o{ SUBJECTS : contains
    SUBJECTS ||--o{ SAMPLES : provides
    TREATMENTS ||--o{ SAMPLE_TIMEPOINTS : administered
    SAMPLES ||--|| CELL_COUNTS : has
    SAMPLE_TIMEPOINTS }|--|| SAMPLES : references

````python
#Code - Python with visualization

from IPython.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd

fig, ax = plt.subplots(figsize=(10, 6))
ax.text(0.1, 0.9, "PROJECTS\n(project_id, name)", bbox=dict(facecolor='white', alpha=0.5))
ax.text(0.4, 0.7, "SUBJECTS\n(subject_id, project_id, condition)", bbox=dict(facecolor='white', alpha=0.5))
ax.text(0.7, 0.5, "SAMPLES\n(sample_id, subject_id, type)", bbox=dict(facecolor='white', alpha=0.5))
ax.text(0.7, 0.3, "CELL_COUNTS\n(count_id, sample_id, counts...)", bbox=dict(facecolor='white', alpha=0.5))
ax.text(0.4, 0.3, "TREATMENTS\n(treatment_id, name)", bbox=dict(facecolor='white', alpha=0.5))
ax.text(0.4, 0.5, "SAMPLE_TIMEPOINTS\n(timepoint_id, sample_id, treatment_id)", bbox=dict(facecolor='white', alpha=0.5))

ax.arrow(0.2, 0.85, 0.15, -0.1, head_width=0.02, head_length=0.02, fc='k', ec='k')
ax.arrow(0.5, 0.65, 0.15, -0.1, head_width=0.02, head_length=0.02, fc='k', ec='k')
ax.arrow(0.5, 0.45, 0.15, 0.0, head_width=0.02, head_length=0.02, fc='k', ec='k')
ax.arrow(0.55, 0.35, 0.1, 0.1, head_width=0.02, head_length=0.02, fc='k', ec='k')
ax.arrow(0.45, 0.35, -0.1, 0.1, head_width=0.02, head_length=0.02, fc='k', ec='k')

ax.set_title("Database Schema Diagram")
ax.axis('off')
plt.show()
````
